```python
# What is the mean and standard deviation across the whole dataset for the year 2017 to 3 decimal places?
mean_2017 = df['Y2017'].mean()
std_2017 = df['Y2017'].std()
print(f'The mean for 2017 is {round(mean_2017,3)}')
print(f'The std for 2017 is {round(std_2017, 3)}')
```

```python
# What is the total number and percentage of missing data in 2014 to 2 decimal places?
total_missing =df['Y2014'].isnull().sum()
percent_missing = df['Y2014'].isnull().sum() * 100 / len(df['Y2014'])
print(f' The total missing data in 2014 is {total_missing}')
print(f' The percentage of  missing data in 2014 is {round(percent_missing, 2)}%')
```
